```python
# Gráfico de Barras População e Religião
fig = px.bar(
    df,
    x='Religião',
    y='População',
    title='Distribuição das principais religiões e populações',
    labels={'Religião', 'População'},
    color='População',
    color_continuous_scale=px.colors.sequential.Plasma
)
fig.update_traces(text=df['População'].astype(str), textposition='outside')
fig.update_layout(xaxis_title='Religião', yaxis_title='População')

fig.show()
```
